```python
import pandas as pd
import jieba
import numpy as np

neg = pd.read_excel('./sentiment/neg.xls',header=None,index =None)
pos = pd.read_excel('./sentiment/pos.xls',header=None,index =None)
pos['mark'], neg['mark'] = 1, 0
pn = pd.concat([pos,neg], ignore_index=True)
pn.columns=['comment', 'mark']
neglen, poslen = len(neg), len(pos)

cw = lambda x: list(jieba.cut(x))
pnWords = [list(jieba.cut(sen)) for sen in pn['comment']] # 保留评论对应的分词结果
# print(pn)

comment = pd.read_excel('./sentiment/sum.xls') # 爬取的淘宝评论
comment = comment[comment['rateContent'].notnull()] # 取文字评论非空的评论
commentWords = [list(jieba.cut(sen)) for sen in comment['rateContent'] ] # 分词
# print(commentWords)
d2v_train = pnWords.extend(commentWords)
# print(d2v_train)

w = [] #将所有词语整合在一起

for wordsList in d2v_train:
    w.extend(wordsList)  # 降维，所有句子合并到一个list
print(w)

countDict = pd.DataFrame(pd.Series(w).value_counts())  # 统计词频
countDict['id']=list(range(1,len(countDict)+1))
# print(countDict)

get_sent = lambda x: list(countDict['id'])
pn['sent'] = pn['words'].apply(get_sent) 
print(pn)

pn.to_excel('./sentiment/pn.xls')

maxlen = 50
```
